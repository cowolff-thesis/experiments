```python
import matplotlib.pyplot as plt
import pandas as pd
import glob
import matplotlib.lines as mlines

# Correcting the label handling and ensuring proper plotting

# Function to plot the data
def plot_experiment_results(files, color_map):
    plt.figure(figsize=(14, 10))

    for file in files:
        df = pd.read_csv(file)
        steps = df['Step']
        values = df['Value']
        linestyle = '--' if 'test' in file.split('/')[-1] else '-'
        label = file.split('/')[-1].replace('.csv', '').replace('_', ' ').capitalize()
        plt.plot(steps, values, linestyle, color=color_map[file.split('/')[-1]], label=label)

    blue_line = mlines.Line2D([], [], color='blue', label='Team with language')
    red_line = mlines.Line2D([], [], color='red', label='Team without language')
    light_blue_ine = mlines.Line2D([], [], color='#21befc', label='Test with language')
    light_red_line = mlines.Line2D([], [], color='#eb7434', label='Test without language')

    plt.xlabel('Step')
    plt.ylabel('Win Rate')
    plt.title('Experiment Results')
    plt.legend(handles=[blue_line, red_line, light_blue_ine, light_red_line])
    plt.show()

# Define file paths and color map
file_pattern = 'data/*.csv'
files = glob.glob(file_pattern)

color_map = {
    'exp1_team_one_wins.csv': 'blue',
    'exp1_team_two_wins.csv': 'red',
    'exp1_test_one_wins.csv': '#21befc',
    'exp1_test_two_wins.csv': '#eb7434',
    'exp2_team_one_wins.csv': 'blue',
    'exp2_team_two_wins.csv': 'red',
    'exp2_test_one_wins.csv': '#21befc',
    'exp2_test_two_wins.csv': '#eb7434'
}

# Plot the results
plot_experiment_results(files, color_map)
```
